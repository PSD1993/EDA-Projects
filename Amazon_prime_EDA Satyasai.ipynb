```python
psd["seasons"].fillna(0,inplace=True)
```

```python
psd.dropna(subset=['imdb_id'], inplace=True)
```

```python
psd['imdb_score'].fillna(round(psd['imdb_score'].mean(),1),inplace=True)
```

```python
psd['imdb_votes'].fillna(0,inplace=True)
```

```python
psd['tmdb_popularity'].fillna(round(psd['tmdb_popularity'].mean(),2), inplace=True)
```

```python
psd['tmdb_score'].fillna(round(psd['tmdb_score'].mean(),1),inplace=True)
```

```python
psd['character'].fillna('unknown',inplace=True)
```

|-------------------------------------- 📊 Box Plot of IMDB Score – Full Visualization Explained 📊------------------------------------------------|

A box plot is used to understand the distribution of numerical data and identify outliers.


🧱 1. Box (rectangle):

This box shows the range from the 25th percentile (Q1) to the 75th percentile (Q3), also known as the IQR (Interquartile Range).

The line inside the box represents the median (50th percentile). In this plot, it is approximately around 6.

Meaning: Half of the IMDB scores fall between 5.5 and 7.5.

📏 2. Whiskers (lines extending from the box):

The left whisker goes up to Q1 − 1.5×IQR

The right whisker goes up to Q3 + 1.5×IQR

Any data points beyond these whiskers are considered outliers.

⚪ 3. Outliers (small circular dots):

Many outliers are visible on the left side (IMDB scores less than 2).

Only a few outliers are seen on the right side (scores greater than 9).

👉 This means:

There are many TV Shows and movies with very low scores (IMDB score < 3), shown on the left.

There are very few high-rated TV Shows and movies (score > 9), shown on the right.

📌 Summary:

Feature	Observation
Median Score	Around 6
Most Scores Between	5.5 to 7.5 (IQR range)
Left Side Outliers	Many – several TV Shows and movies with very low scores
Right Side Outliers	Few – only a few TV Shows and movies with very high scores

####### Count plot of genres

```python
# Step 1: Count value of genres
genre_counts = psd['genres'].value_counts()

# Step 2: Filter genres with count > 1000
psd_genres = genre_counts[genre_counts > 1000].index

#This line filters the original DataFrame psd to keep only those rows where the value in the 'genres' column is present in psd_genres.
psd_Data = psd[psd['genres'].isin(psd_genres)] 
plt.figure(figsize=(12, 6))

dp = sns.countplot(y='genres', data=psd_Data, palette='bright', edgecolor='black')
for container in dp.containers:
    dp.bar_label(container, color='blue', fontsize=10)
    
plt.title('Count plot for Genres', color='orange', fontsize=20)
plt.xlabel('Count', color='blue', fontsize=14)
plt.ylabel('Genres', color='blue', fontsize=14)
plt.tight_layout() # It is use for Automatically adjusts the subplot parameters so that:
plt.show()
```

. This horizontal bar chart displays the frequency of different genre combinations in the dataset.
.Each bar represents a unique genre or a combination of genres, and the number at the end of each bar is the count of records (e.g., movies or shows)      that fall into that genre.

Key Insights:

The dataset shows a strong preference for genres that involve drama, comedy, and romance, either standalone or in combination. Thriller, horror, and documentary genres are also present but to a lesser extent. This likely reflects audience demand or the content strategy of the platform from which the data is sourced.

```python
#Count plot of average rating for each age certification


# Step 1: Group by age certification and calculate average rating
avg_rating_df = psd.groupby('age_certification')['imdb_score'].mean().reset_index()

# Step 2: Sort (optional, for better visualization)
avg_rating_df = avg_rating_df.sort_values(by='imdb_score', ascending=False)

# Step 3: Plot
plt.figure(figsize=(10, 5))
ax = sns.barplot(x='age_certification', y='imdb_score', data=avg_rating_df, palette='bright6', edgecolor='black')

# Step 4: Add labels on top of each bar
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', fontsize=10, color='black')

# Step 5: Titles and labels
plt.title('Average Rating by Age Certification', fontsize=16, color='darkblue')
plt.xlabel('Age Certification', fontsize=12, color='red')
plt.ylabel('IMDB Score', fontsize=12, color='red')

plt.tight_layout()
plt.show()
```

PIE CHART

Pie chart is used to visualize the proportions of different categories within a whole dataset. It's circular statistical graph, where each slice of the pie represents a category, and the size of the slice is proportional to the category's contribution to the overall data.

```python
####Pie chart of type(Movie or TV show)
# Count the values once and store it
type_counts = psd['type'].value_counts()

plt.figure(figsize=(8,6))
plt.pie(type_counts, labels=type_counts.index, autopct='%.2f%%',explode =[0.2, 0],shadow=True,startangle=10 )
plt.title('Pie chart of Type (Movie or TV Show)', color='brown')
plt.legend(type_counts.index, title="Type")
plt.show()
```

```python
#Violin Plot of imdb_score by type
plt.figure(figsize=(7, 4))
sns.violinplot(x='type', y='imdb_score', data=psd)
plt.title('Violin Plot of IMDB Score by Type',color='red',fontsize=14)
plt.xlabel('Type',color='red',fontsize=14)
plt.ylabel('IMDB Score', color='red',fontsize=14)
plt.show()
```
